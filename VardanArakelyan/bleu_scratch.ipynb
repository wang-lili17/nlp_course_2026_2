```python
def get_n_grams(list, n):
    res = {}
    for i in range(len(list) - n + 1):
        n_gram = tuple(list[i:i + n])
        if n_gram not in res:
            res[n_gram] = 1
        else:
            res[n_gram] += 1
    return res
lists = ['catr', 'is', 'done']
n = 2
print(get_n_grams(lists, n))

```

```python
def clipped_precision(hyp, ref, n):
    hyp_ngrams = get_n_grams(hyp, n)
    ref_ngrams = get_n_grams(ref, n)
    num = []
    for hyp_ngram in hyp_ngrams:
        if hyp_ngram in ref_ngrams:
            num.append(min(hyp_ngrams[hyp_ngram], ref_ngrams.get(hyp_ngram,0)))
    sum_num = sum(num)
    sum_den = 0
    for key, val in hyp_ngrams.items():
        sum_den += val
    if sum_den == 0:
        return 0.0
    res = sum_num / sum_den

    return res
```

```python
import math
def brevity_penalty(hyp, ref):
    len_hyp = len(hyp)
    len_ref = len(ref)
    if len_hyp >= len_ref:
        return 1
    else:
        return math.exp(1 - (len_ref / len_hyp))
```

```python
def bleu(hyp, ref, max_n = 4):
    n = list(range(1, max_n + 1))
    clip_res = []
    for i in n:
        clip_res.append(clipped_precision(hyp, ref, i))
    if 0.0 in clip_res:
        return 0.0
    log_sum = 0
    for i in clip_res:
        log_sum += math.log(i) * (1/4)
    sum_clip = math.exp(log_sum)

    return brevity_penalty(hyp, ref) * sum_clip

```

```python
hyp = "the match was postponed because of the snow".split()
ref = "the match was postponed because it was snowing".split()
print(bleu(hyp, ref, 1))
print(bleu(hyp, ref, 2))
print(bleu(hyp, ref))
# Expected: ~0.516
```
